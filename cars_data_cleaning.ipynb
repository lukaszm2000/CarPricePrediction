```python
def convert_price(x):
    x = x.replace('₹', '').replace(',', '')  
    if 'Crore' in x:
        return float(x.replace('Crore', '')) * 100
    elif 'Lakh' in x:
        return float(x.replace('Lakh', ''))
    else:
        return float(x) / 100000

df['resale_price_Lakh'] = df['resale_price'].apply(convert_price)
```

```python
#wartosc 0.755 zostala wybrana jako srednia gestosc dla benzyny i diesla w celu przeksztalcenia jednostki km/kg(kilometers per kilogram) na kmpl(kilometers per liter)

def convert_mileage(x):
    try:
        if x.endswith('km/kg'):
            return str(2*round(1/2*float(0.755*str(x).rstrip(' km/kg')))) 
        elif x.endswith('kmpl'):
            return str(2*round(1/2*float(str(x).rstrip(' kmpl'))))
    except:
        return 'None'
```

```python
df['mileage']
```

```python
df['mileage'] = df['mileage'].apply(convert_mileage)
```
