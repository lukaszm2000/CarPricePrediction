```python
def convert_price(x):
    x = x.replace('₹', '').replace(',', '').strip()  
    if 'Crore' in x:
        return float(x.replace('Crore', '')) * 100
    elif 'Lakh' in x:
        return float(x.replace('Lakh', ''))
    else:
        return float(x) / 100000

df['resale_price_Lakh'] = df['resale_price'].apply(convert_price)
```

```python
def transform_max_power(x):
    try:
        x = x.lower().replace('bhp', '').replace('ps', '').strip()  
        if x:  
            return str(int(float(x) / 10) * 10)  
        else:
            return 'None'
    except:
        return 'None'
```

```python
def convert_kmpl_to_kmkg(value, fuel_type):
    if 'kmpl' in value:  
        value = float(value.replace(' kmpl', '')) 
        if fuel_type == 'Petrol':
            return value / 0.75  
        elif fuel_type == 'Diesel':
            return value / 0.85  
        elif fuel_type == 'LPG':
            return value / 0.55  
        else:
            return value  
    else:
        return float(value.replace(' km/kg', ''))
    
df['mileage'] = df.apply(lambda row: convert_kmpl_to_kmkg(row['mileage'], row['fuel_type']) 
                         if 'kmpl' in row['mileage'] else float(row['mileage'].replace(' km/kg', '')), axis=1)
```
